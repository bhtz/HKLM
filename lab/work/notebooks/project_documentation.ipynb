```python
### Import modules

import pandas as pd
from sqlalchemy import *
from eralchemy import render_er
from IPython.display import Image
```

```python
### Database metadata

CONNECTION_STRING = 'postgresql://postgres:root@host.docker.internal:5432/postgres'

engine = create_engine(CONNECTION_STRING)
metadata = MetaData(engine)
inspector = inspect(engine)
```

### Exposed tables :

```python
query = 'SELECT table_schema, table_name FROM hdb_catalog.hdb_table WHERE table_schema NOT IN (\'hdb_catalog\', \'information_schema\')'
df = pd.read_sql_query(query, con=engine)
df.head()
```

### Actions :

```python
query = 'SELECT action_name, comment FROM hdb_catalog.hdb_action'
df = pd.read_sql_query(query, con=engine)
df.head()
```

### Remote schemas :

```python
query = 'SELECT name FROM hdb_catalog.remote_schemas'
df = pd.read_sql_query(query, con=engine)
df.head()
```
